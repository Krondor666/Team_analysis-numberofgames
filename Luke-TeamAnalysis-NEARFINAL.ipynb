```python
# Need a function that pulls out only a certain teams data:

def team_data(team): # probably by index...
    df = dat.loc[(dat.ix[:,3] == team) |  (dat.ix[:,4] == team) ]
    return(df)
```

```python
# win ratio metric:

def win_ratio(team):
    # get team data...
    dato = team_data(team).reset_index(drop=True)
    #print('The subset data length is:', len(dato))
    # then iterate over every row:
    WL = []
    #print(np.shape(data)[0])
    for row in range(len(dato)):
        # print(dat.ix[row,5])    # this here jsut for testing...
        if (dato.ix[row,5]) == team : # this is if they are the winning team - name is in the winning team col...
            WL.append(1)
            
        else:
            WL.append(0)
            
    #return(WL)
    #return(np.sum(WL)/len(WL), np.std(WL)) # this is the Win ratio metric... and variance (STD)
    return(WL)  
```

```python
# Need a function that pulls out only a certain player data:

def player_data(player): # probably by index...
    df = dat.loc[(dat.ix[:,11] == player) |  (dat.ix[:,17] == player)|  (dat.ix[:,23] == player)|  (dat.ix[:,29] == player)|  
                 (dat.ix[:,35] == player)|  (dat.ix[:,41] == player)|  (dat.ix[:,47] == player)|  (dat.ix[:,53] == player)| 
                 (dat.ix[:,59] == player) |  (dat.ix[:,65] == player)]
    return(df)

player_data('LTN')

#Heh, actually works...
```

```python
def player_team(player): # match is the match # as integer

    dato = player_data(player).reset_index(drop=True) 
    
    side = []
    
    for row in range(len(dato)):
        if( (dato.ix[row,11] == player) |  (dato.ix[row,17] == player) |  
          (dato.ix[row,23] == player)  |  (dato.ix[row,29] == player) |  
          (dato.ix[row,35] == player) ):
            side.append(dato.ix[row,'Team 1'])
            
        elif( (dato.ix[row,41] == player) |  (dato.ix[row,47] == player) | 
            (dato.ix[row,53] == player)  |  (dat.ix[row,59] == player)  |  
            (dato.ix[row,65] == player) ):
             side.append(dato.ix[row,'Team 2'])
            
    return(side)
#player_team('GeT_RiGhT')

#This returns a list of the team that the called 'player' was on for each match.
#Actually so FUCKING pumped I got this to work
```

```python
def player_unique_andfreq(player):

    a=player_team(player)

    b=[(k, len(list(v))) for k, v in groupby(a)]

    return b
#player_unique_andfreq('GeT_RiGhT') #this is a list, want to turn into a dataframe

def dataframe(player):
    
    frame=pd.DataFrame(player_unique_andfreq(player), columns=["Team", "Number of games",])
    
    return(frame)
dataframe('GeT_RiGhT')

#This is in order, with the most recent roster at the bottom.
#chickayeaboiiii
```
